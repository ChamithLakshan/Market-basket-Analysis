```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv("Online_Retail.csv",encoding='latin1')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['ï»¿InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['ï»¿InvoiceNo'].astype('str')
df=df[~df['InvoiceNo'].str.contains('C')]
```

```python
df.head()
```

```python
df.shape
```

```python
basket =(df[df['Country'] == "France"]
        .groupby(['InvoiceNo','Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
basket.shape
```

```python
 def encode_units(x):
        if x<=0:
            return 0
        else:
            return 1

        
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

```python
frequent_item_sets = apriori(basket_sets,min_support=0.10, use_colnames=True)
```

```python
rules = association_rules(frequent_item_sets,metric="lift",min_threshold=1)
rules.head()
```

```python
rules[(rules['lift'] >= 1) &
       (rules['confidence']>= 0.8)]
```

```python
basket_uk =(df[df['Country'] == "Germany"]
        .groupby(['InvoiceNo','Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
```

```python
basket_sets_uk = basket_uk.applymap(encode_units)
frequent_item_sets = apriori(basket_sets_uk,min_support=0.07, use_colnames=True)
```

```python
rules_uk = association_rules(frequent_item_sets,metric="lift",min_threshold=1)
rules_uk
```

```python
rules_uk[(rules_uk['lift'] >= 1) &
       (rules_uk['confidence']>= 0.8)]
```
